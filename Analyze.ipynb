```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt

directory = "C:/Users/Alket/Code/Alluvione/data" # use your path

# Specify the directory where the CSV files are located

# Get a list of all CSV file paths in the directory
csv_files = glob.glob(directory + '/*.csv')

# Initialize an empty list to store individual dataframes
dataframes = []

# Read each CSV file and append its dataframe to the list
for file in csv_files:
    df = pd.read_csv(file)
    dataframes.append(df)

# Concatenate all dataframes into a single dataframe
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
combined_df
```

```python
plt.figure(figsize = (20, 5))
plt.plot(combined_df["DAILY_PREC"])
plt.show()
```

```python
media_mobile = combined_df["DAILY_PREC"].rolling(90).mean()
```

```python
plt.figure(figsize = (20, 5))
plt.plot(media_mobile)
plt.show()
```
