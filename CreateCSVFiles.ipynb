```python
import os
import csv

def combine_csv_files(folder_path, output_folder):
    # Get a list of all subfolders within the given folder
    subfolders = [f.path for f in os.scandir(folder_path) if f.is_dir()]

    for subfolder in subfolders:
        # Get the name of the subfolder
        subfolder_name = os.path.basename(subfolder)

        # Create a new CSV file for the subfolder
        csv_file_path = os.path.join(output_folder, f"{subfolder_name}.csv")
        with open(csv_file_path, mode='w', newline='') as csv_file:
            writer = csv.writer(csv_file)

            # Find all CSV files within the subfolder
            csv_files = [f.path for f in os.scandir(subfolder) if f.is_file() and f.name.endswith('.csv')]

            # Combine the CSV files into the subfolder's CSV file
            for csv_file_path in csv_files:
                with open(csv_file_path, mode='r') as input_csv:
                    reader = csv.reader(input_csv)
                    writer.writerows(reader)

        print(f"Created CSV file for '{subfolder_name}': {csv_file_path}")


```

```python
# Provide the path to the folder containing the subfolders
folder_path = "C:/Users/Alket/Code/Alluvione"

# Provide the path to the output folder where the main CSV files will be written
output_folder = "C:/Users/Alket/Desktop/alldata_precisp"

combine_csv_files(folder_path, output_folder)

```

```python
import pandas as pd

def drop_rows_every_n(dataframe, n):
    rows_to_drop = list(range(n-1, len(dataframe), n))
    dataframe = dataframe.drop(rows_to_drop)
    return dataframe

# Example usage:
# Create a sample DataFrame
df = pd.DataFrame({'A': range(1, 11), 'B': range(11, 21)})

# Drop one row every 3 rows
df = drop_rows_every_n(df, 3)

# Print the resulting DataFrame
print(df)
```
